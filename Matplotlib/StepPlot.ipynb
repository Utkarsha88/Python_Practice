# Step Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.step(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,2,1,3]
plt.step(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,2,1,3]
plt.step(x,y)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,2,1,3]
plt.step(x,y)
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("Python")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,2,1,3]
plt.step(x,y,color="r",marker="o",markersize="10",mfc="g",label="python")
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("Python")
plt.legend()
plt.grid()
plt.show()
```
