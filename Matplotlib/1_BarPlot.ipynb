# Bar Plot

```python
import matplotlib.pyplot as plt
x=["Python","c","c++","Java"]
y=[85,70,60,82]
plt.xlabel("<------Languages-------->")
plt.ylabel("Population")
plt.title("Language Population Bar")
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["Python","c","c++","Java"]
y=[85,70,60,82]
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Population",fontsize=15)
plt.title("Language Population Bar",fontsize =20)
plt.bar(x,y,width=0.2,color="y")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["Python","c","c++","Java"]
y=[85,70,60,82]
z=[50,25,50,50]
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Population",fontsize=15)
plt.title("Language Population Bar",fontsize =20)
plt.bar(x,y,width=0.2,color="r",edgecolor="black",label="Utkarsh") 
plt.bar(x,z,width=0.2,color="y",edgecolor="black",label="Gupta") 
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["Python","c","c++","Java"]
y=[85,70,60,82]
z=[50,25,50,50]
width=0.4
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Languages",fontsize=15)
plt.ylabel("Population",fontsize=15)
plt.title("Language Population Bar",fontsize =20)

plt.bar(p,y,width,color="r",edgecolor="black",label="Utkarsh") 
plt.bar(p1,z,width,color="y",edgecolor="black",label="Gupta") 

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["Python","c","c++","Java"]
y=[85,70,60,82]
z=[50,25,50,50]
width=0.4
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Languages",fontsize=15)
plt.ylabel("Population",fontsize=15)
plt.title("Language Population Bar",fontsize =20)

plt.bar(p,y,width,color="r",edgecolor="black",label="Utkarsh") 
plt.bar(p1,z,width,color="y",edgecolor="black",label="Gupta") 

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["Python","c","c++","Java"]
y=[85,70,60,82]
z=[50,25,50,50]
width=0.4
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Languages",fontsize=15)
plt.ylabel("Population",fontsize=15)
plt.title("Language Population Bar",fontsize =20)

plt.bar(p,y,width,color="r",edgecolor="black",label="Utkarsh") 
plt.bar(p1,z,width,color="y",edgecolor="black",label="Gupta") 

plt.xticks(p+width/2,x,rotation=30)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["Python","c","c++","Java"]
y=[85,70,60,82]
z=[50,25,50,50]
width=0.4
p=np.arange(len(x))
p1=[j+width for j in p]


plt.ylabel("Languages",fontsize=15)
plt.xlabel("Population",fontsize=15)
plt.title("Language Population Bar",fontsize =20)

plt.barh(x,y,width,color="r",edgecolor="black",label="Utkarsh") 
plt.barh(p1,z,width,color="y",edgecolor="black",label="Gupta") 

plt.yticks(p+width/2,x,rotation =10)
plt.legend()
plt.show()
```
