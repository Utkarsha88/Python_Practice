# Stack Plot ,Area Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area=[1,2,3,4,5]

plt.stackplot(x,area)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area=[2,5,6,3,4]

plt.stackplot(x,area)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area=[2,5,6,3,4]
area2=[4,5,9,4,7]
area3=[5,6,9,3,5]
l=["area","area2","area3"]
color=["r","g","b"]

plt.stackplot(x,area,area2,area3,labels=l,colors=color)
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area=[2,5,6,3,4]
area2=[4,5,9,4,7]
area3=[5,6,9,3,5]
l=["area","area2","area3"]
color=["r","g","b"]

plt.stackplot(x,area,area2,area3,labels=l,colors=color,baseline="sym")
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area=[2,5,6,3,4]
area2=[4,5,9,4,7]
area3=[5,6,9,3,5]
l=["area","area2","area3"]
color=["r","g","b"]

plt.stackplot(x,area,area2,area3,labels=l,colors=color,baseline="wiggle")
plt.title("python")
plt.xlabel("xaxis")
plt.ylabel("yaxis")
# plt.grid()

plt.legend(loc=2)
plt.show()
```
