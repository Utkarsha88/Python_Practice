# PieChart

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.1,0.0,0.2,0.0]
c=["r","b","y","g"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
x1=[40,30,10,55]
y=["C","C++","Java","Python"]

plt.pie(x,labels=y,radius=1.5)
cr=plt.Circle(xy=(0,0),radius=1,facecolor="white")
plt.gca().add_artist(cr)
plt.show()
```
