# Stem Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[8,6,5,2,3,4]
plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[8,6,5,2,3,4]
plt.stem(x,y,linefmt=":")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[8,6,5,2,3,4]
plt.stem(x,y,linefmt=":",markerfmt="+")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[8,6,5,2,3,4]
plt.stem(x,y,linefmt=":",markerfmt="r+")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[8,6,5,2,3,4]
plt.stem(x,y,linefmt=":",markerfmt="+",bottom=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[8,6,5,2,3,4]
plt.stem(x,y,linefmt=":",markerfmt="+",bottom=9)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[8,6,5,2,3,4]
plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=9,basefmt="g",label="python",use_line_collection=False)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[8,6,5,2,3,4]
plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt="g",label="python",orientation="horizontal")
plt.legend()
plt.show()
```
