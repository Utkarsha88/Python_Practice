# Axis Matplotlib Plots

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[3,1,4,5,2]
plt.plot(x,y)



plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,1,4,5,2]
plt.plot(x,y)

plt.xticks(x)
plt.yticks(y)

plt.show()
```

```python
x=[1,2,3,4]
y=[3,1,4,5]
plt.plot(x,y)

plt.xticks(x , labels=["python","C","C++","Java"])
plt.yticks(y)

plt.show()
```

```python
x=[1,2,3,4]
y=[3,1,4,5]
plt.plot(x,y)

# plt.xticks(x , labels=["python","C","C++","Java"])
# plt.yticks(y)
plt.xlim(0,10)

plt.show()
```

```python
x=[1,2,3,4]
y=[3,1,4,5]
plt.plot(x,y)

# plt.xticks(x , labels=["python","C","C++","Java"])
# plt.yticks(y)
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```python
x=[1,2,3,4]
y=[3,1,4,5]
plt.plot(x,y)


plt.axis([0,10,0,10])

plt.show()
```
